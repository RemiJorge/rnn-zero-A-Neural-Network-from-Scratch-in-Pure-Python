```python
from nnfs.datasets import spiral_data
import numpy as np
import nnfs
nnfs.init()
import matplotlib.pyplot as plt
X, y = spiral_data(samples=200, classes=3)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='brg')
plt.show()
```

```python
from src.neural_network import NeuralNetwork, Optimizer_Adam, ReLU, DenseLayer

nn = NeuralNetwork(
    layers=[
        DenseLayer(2, 64), 
        ReLU(),
        DenseLayer(64,3)
        ],
    optimizer=Optimizer_Adam(learning_rate=0.02, decay=1e-5))
```

```python
for epoch in range(10001):
    loss, acc, val_loss, val_acc = nn.train(X, y)
    if epoch%100 == 0:
        print(f"epoch: {epoch}, acc: {acc:.3f}, val: {val_acc:.3f}, loss: {loss:.3f}, learning rate: {nn.get_current_learning_rate():.3f}")
```
